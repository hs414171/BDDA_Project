```python
messages = X.copy()
labels = y.copy()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
corpus = []
classlab = []
for i in range(0, len(messages)):
    try:
        review = re.sub('[^a-zA-Z]', ' ', messages['CONTENT'][i])
        review = review.lower()
        review = review.split()
        
        review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
        review = ' '.join(review)
        corpus.append(review)
        
    except Exception as e:
        print(f"Error occurred for index {i}. Skipping...")
        continue
```

```python
len(corpus)
```

```python
corpus
```

```python
voc_size = 5000
onehot_repr = [one_hot(words,voc_size) for words in corpus]
onehot_repr
```

```python
max_length = 0
for sublist in onehot_repr:
    sublist_length = len(sublist)
    if sublist_length > max_length:
        max_length = sublist_length
        
print(f"The maximum length of a sublist is: {max_length}")
```

```python
embedded_docs = pad_sequences(onehot_repr,padding='pre',maxlen=max_length)
```

```python
embedded_docs
```

```python
embedding_vector_features=320
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length = max_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size=0.33,random_state=55)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
for i in range(0,len(y_pred)):
    if y_pred[i]>0.5:
        y_pred[i] = 1
    else:
        y_pred[i]=0
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
